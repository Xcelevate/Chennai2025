```python
import random
print("Welcome to Rock-Paper-Scissors!")
user = input("Enter your choice (rock/paper/scissors): ").lower()
if not user.isalpha():
    print("Invalid input. Please choose rock, paper, or scissors.")
else:
    n = random.randint(1,3)
    if n == 1:
        computer = "rock"
    elif n == 2:
        computer = "paper"
    else:
        computer = "scissors"
    if user in ["rock", "paper", "scissors"]:
        print("You chose:", user)
        print("Computer chose:", computer)
        if user == computer:
            print("It's a tie!")
        elif (user == 'rock' and computer == 'scissors') or (user == 'paper' and computer == 'rock') or (user == 'scissors' and computer == 'paper'):
            print("You win!")
        else:
            print("Computer wins!")
    else:
        print("Give only rock, paper, or scissors!")
```

```python
print("Welcome to Treasure Island!")
print("Your mission is to find the treasure.")
print("Type left or right")
c1 = input().lower()
if c1 == "right":
    print("You fell into a hole. Game Over.")
elif c1 == "left":
    print("You've come to a lake. There is an island in the middle of the lake.")
    print("Type wait to wait for a boat. Type swim to swim across.")
    c2 = input().lower()
    if c2 == "wait":
        print("You arrive at the island unharmed. There is a house with 3 doors.")
        print("One red, one yellow, and one blue. Which colour do you choose?")
        c3 = input().lower()
        if c3 == "red":
            print("It's a room full of fire. Game Over.")
        elif c3 == "blue":
            print("You enter a room of beasts. Game Over.")
        elif c3 == "yellow":
            print("You found the treasure! You Win! ðŸ†")
        else:
            print("Invalid choice! Choose red, yellow, or blue.")
    elif c2 == "swim":
        print("You get attacked by an angry trout. Game Over.")
    else:
        print("Invalid input! Choose 'wait' or 'swim'.")
else:
    print("Invalid input! Please choose only 'left' or 'right'.")
```
