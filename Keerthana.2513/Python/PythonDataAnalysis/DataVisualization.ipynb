#### Basic Line Plot Create a line plot showing the total revenue per month for the year 2024.X-axis: Month  Y-axis: Total Revenue
#### Add appropriate labels and title Add a grid for better readability

```python
total_revenue=df.groupby('Month').agg({'Revenue':'sum'})
```

```python
total_revenue
```

```python
# Create line plot 

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(total_revenue.index, total_revenue.values,
marker='o', linewidth=2, markersize=8, color='green')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.title('Monthly Revenue Trend - 2024', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.4)
plt.xticks(range(1, 13))
plt.tight_layout()
plt.show()
```

```python
function 
plot-helps us draw the graph it helps us determine the color and the index and values
xlabel-helps us label the x axis
ylabel-helps us label the y axis
title-helps us put the title to the image
grid-helps us put the checkboxes created
xticks-no of the elements that we give for x axis
tight_layout-the rectangle box that we see is tight_layout
show-it is similiar t0
```
