# 3.Duplicates

## Feature Engineering

#### a) Extract the month and year from the Date column into new columns.

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Month']=df['Date'].dt.month
```

```python
df['Year']=df['Date'].dt.year
```
